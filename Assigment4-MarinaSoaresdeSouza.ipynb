```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
type(tips)
```

```python
tips.head()
```

```python
dcounts = tips['day'].value_counts()
day_order = ['Thur', 'Fri', 'Sat', 'Sun']
dcounts = dcounts.reindex(day_order, fill_value=0)
dcounts
print(dcounts)
```

Pandas

```python
dcounts.plot(kind='bar')
plt.title('Transactions by Day')
plt.xlabel('Day')
plt.ylabel('Number of Transactions')
plt.yticks([10,20,30,40,50,60,70,80,90])
plt.show()
```

Seanborn

```python
plot1=sns.countplot(x='day', data=tips)
plt.title('Transactions by Day')
plt.xlabel('Day')
plt.ylabel('No of Transactions')
plt.yticks([10,20,30,40,50,60,70,80,90])
```
